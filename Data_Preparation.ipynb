```python
import pandas as pd
import numpy as np
import os
```

```python
data_dir = 'data/'
```

```python
df_train1 = pd.read_csv(os.path.join(data_dir, 'train_1.csv'))
```

```python
df_train2 = pd.read_csv(os.path.join(data_dir, 'train_2.csv'))
```

```python
df_train1.head()
```

```python
df_train2.head()
```

```python
split_pages = df_train2['Page'].apply(lambda x: x.split('_')).tolist()
merged_name_pages = []
for split_page in split_pages:
    merged_name_page = []
    merged_name_page.append('_'.join(split_page[:-3]))
    merged_name_page.extend(split_page[-3:])
    merged_name_pages.append(merged_name_page)
df_split_pages = pd.DataFrame(merged_name_pages, columns=['Name', 'Project', 'Access', 'Agent'])
```

```python
df_split_pages.head()
```

```python
columns = df_train2.columns.values
df_train_final = pd.concat([df_train2[columns[0]], df_split_pages, df_train2[columns[1:]]], axis=1)
```

```python
df_train_final.head()
```

```python
df_train_sample = df_train_final.dropna().sample(n=100, random_state=42).reset_index(drop=True)
```

```python
df_train_sample.head()
```

```python
len(df_train_sample)
```

```python
df_train_sample.to_csv(os.path.join(data_dir, 'train_clean_sample.csv'), index=False)
```

```python
%%time
df_train_final.to_csv(os.path.join(data_dir, 'train_final.csv'), index=False)
```
