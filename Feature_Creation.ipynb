```python
from fastai.tabular import * 
import pandas as pd
import numpy as np
import os
```

```python
print(fastai.__version__)
```

```python
data_dir = 'data/'
df_train = pd.read_csv(os.path.join(data_dir, 'train_clean_sample.csv'))
```

```python
df_train.head()
```

```python
df_train = df_train.melt(id_vars = df_train.columns[:5])#.set_index('variable')
df_train.head()
```

```python
df_train.variable.u
```

```python
df_train_lag = df_train.groupby(['Page']).apply(lambda x: (x
                                            .sort_values('variable', ascending=True))['value'].shift(62)).reset_index().set_index('level_1')
df_train_lag
df_train_lag.columns = ['Page_y', 'lag_value']
df_train = df_train.reset_index().merge(df_train_lag, left_index = True, right_index = True)
```

```python
df_train.head()
```

```python
add_datepart(df_train, "variable", drop=False)
```

```python
df_train.to_csv(os.path.join(data_dir, 'train_with_features.csv'), index=False)
```
